# Healthcare Claims Medallion Architecture - Setup

This notebook contains **shared configuration** for the Healthcare Claims Medallion Pipeline.

## Architecture Layers:

### 🥉 Bronze (Source)
- **Catalog/Schema**: `healthverity_claims_sample_patient_dataset_hr.hv_claims_sample`
- **Tables**: medical_claim, diagnosis, procedure, provider, enrollment
- **Description**: Raw source data as-is

### 🥈 Silver Layer
- **Catalog/Schema**: `hls_amer_catalog.hcspd_hr_silver_claims`
- **Transformations**: Data cleansing, deduplication, standardization, data quality checks
- **Metadata**: Added processed_timestamp and data_quality_flag columns

### 🥇 Gold Layer
- **Catalog/Schema**: `hls_amer_catalog.hcspd_hr_gold_claims`
- **Analytics Tables**: 
  - `patient_summary` - Patient demographics and enrollment analytics
  - `claims_analytics` - Claims aggregated by date, location, procedure
  - `cost_analytics` - Financial analysis of charges vs allowed amounts

```python
# Import required libraries
from pyspark.sql import functions as F
from pyspark.sql.types import *
from datetime import datetime

# Configuration
BRONZE_CATALOG = "healthverity_claims_sample_patient_dataset_hr"
BRONZE_SCHEMA = "hv_claims_sample"
TARGET_CATALOG = "hls_amer_catalog"
SILVER_SCHEMA = "hcspd_hr_silver_claims"
GOLD_SCHEMA = "hcspd_hr_gold_claims"

print(f"Source: {BRONZE_CATALOG}.{BRONZE_SCHEMA}")
print(f"Silver: {TARGET_CATALOG}.{SILVER_SCHEMA}")
print(f"Gold: {TARGET_CATALOG}.{GOLD_SCHEMA}")
```

```python
# Create Silver and Gold schemas in target catalog

spark.sql(f"""
    CREATE SCHEMA IF NOT EXISTS {TARGET_CATALOG}.{SILVER_SCHEMA}
    COMMENT 'Silver layer: Cleaned and standardized healthcare claims data'
""")

spark.sql(f"""
    CREATE SCHEMA IF NOT EXISTS {TARGET_CATALOG}.{GOLD_SCHEMA}
    COMMENT 'Gold layer: Aggregated analytics for healthcare claims'
""")

print(f"✓ Schema ready: {TARGET_CATALOG}.{SILVER_SCHEMA}")
print(f"✓ Schema ready: {TARGET_CATALOG}.{GOLD_SCHEMA}")
```
