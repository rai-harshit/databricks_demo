```python
from databricks.sdk import WorkspaceClient
import json

w = WorkspaceClient()

IMPORT_FILE = "/Workspace/Users/harshit.rai.usa@gmail.com/genie_space_export.json"
TARGET_WAREHOUSE_ID = "5a0ae21a29c598a3"

export_data = json.loads(dbutils.fs.head(IMPORT_FILE, 1000000))

# Create space with everything
new_space = w.api_client.do(
    "POST",
    "/api/2.0/genie/spaces",
    body={
        "title": export_data['title'],
        "description": export_data.get('description'),
        "warehouse_id": TARGET_WAREHOUSE_ID,
        "serialized_space": export_data.get('serialized_space')
    }
)

print(f"Created space: {new_space['space_id']}")
```
